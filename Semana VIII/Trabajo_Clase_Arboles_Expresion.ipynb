## Ejercicio 1

**Expresión a resolver:** `(7 + 2) * (5 - 3)`


### a. Construcción del Árbol


```python
# Escribe aquí el código para construir el árbol
class Nodo: # Definición de la clase Nodo
    def __init__(self, valor):
        self.valor = valor
        self.izq = None
        self.der = None

# Definición de la función para construir el árbol
def construir_arbol(): 
    raiz = Nodo('*') 
    raiz.izq = Nodo('+')
    raiz.der = Nodo('-')
    raiz.izq.izq = Nodo(7)
    raiz.izq.der = Nodo(2)
    raiz.der.izq = Nodo(5)
    raiz.der.der = Nodo(3)
    return raiz
```

### b. Visualización con `networkx`


```python
# Escribe aquí el código para graficar el árbol
import matplotlib.pyplot as plt
import networkx as nx

G = nx.DiGraph()

edges = [
    ('*', '+'),
    ('*', '-'),
    ('+', '7'),
    ('+', '2'),
    ('-', '5'),
    ('-', '3')
]

G.add_edges_from(edges)

pos = {
    '*': (0, 2),
    '+': (-1.5, 1),
    '-': (1.5, 1),
    '7': (-2.2, 0),
    '2': (-0.8, 0),
    '5': (0.8, 0),
    '3': (2.2, 0)
}

plt.figure(figsize=(10, 5))
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=2000,
        font_size=14, font_weight='bold', arrows=True)
plt.title("Árbol de expresión: (7 + 2) * (5 - 3)")
plt.axis('off')
plt.show()
```

### c. Evaluación de la Expresión


```python
# Escribe aquí el código para evaluar la expresión
def evaluar(nodo):
    if isinstance(nodo.valor, int):
        return nodo.valor
    izq = evaluar(nodo.izq)
    der = evaluar(nodo.der)
    if nodo.valor == '+':
        return izq + der
    elif nodo.valor == '-':
        return izq - der
    elif nodo.valor == '*':
        return izq * der
    elif nodo.valor == '/':
        return izq / der

arbol = construir_arbol()
resultado = evaluar(arbol)
resultado
```
